```python
x = df.drop('target',axis=1)
y = df['target']
print(type(x),x.shape)
print(type(y),y.shape)
```

<b>Preprocessing the Data using Standard Scaler:

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit_transform(x)
```

<b>Spliting the Data into Train and Test :

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.30, random_state=22)
print('x_train:',x_train.shape)
print('x_test:',x_test.shape)
print('y_train:',y_train.shape)
print('y_test:',y_test.shape)
```

<b>Creating a function to compute  classification_report,  confusion_matrix,  accuracy_score:

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
def eval_metrics(ytest,ypred):
    acc = accuracy_score(ytest,ypred)
    print('accuracy_score',acc)
    print(confusion_matrix(ytest,ypred))
    print(classification_report(ytest,ypred))
    
def mscore(model):
    print('Training Score',model.score(x_train,y_train))
    print('Testing Score',model.score(x_test,y_test))
```

Build The Models

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier, AdaBoostClassifier
```

<b>logistic Regression Model

```python
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

```python
# Model score 
mscore(lr)
```

```python
# Prediction values on y
ypred_lr = lr.predict(x_test)
print(ypred_lr)
```

```python
# Evaltion Metrics
eval_metrics(y_test,ypred_lr)
```

<b>KNeighbors Classifier Model

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train,y_train)
```

```python
# Model score
mscore(knn)
```

```python
# Prediction values on y
ypred_knn = knn.predict(x_test)
print(ypred_knn)
```

```python
# Evaltion Metrics
eval_metrics(y_test,ypred_knn)
```

<b>Decision Tree Classifier Model

```python
dt = DecisionTreeClassifier(criterion='entropy',max_depth=15,min_samples_split=22)
dt.fit(x_train,y_train)
```

```python
mscore(dt)
```

```python
ypred_dt = dt.predict(x_test)
print(ypred_dt)
```

```python
eval_metrics(y_test,ypred_dt)
```

```python
cn = ['Yes','No']
fn = x_train.columns
print(cn)
print(fn)
```

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(20,20))
plot_tree(dt,class_names=cn,feature_names=fn,filled=True)
plt.show()
```

<b>Ensemble as Ada Boost Classifier Model

```python
ada = AdaBoostClassifier(base_estimator=dt,n_estimators=1)
ada.fit(x_train,y_train)
```

```python
mscore(ada)
```

```python
ypred_ada = ada.predict(x_test)
print(ypred_ada)
```

```python
eval_metrics(y_test,ypred_ada)
```

<b>Ensemble the Bagging Classifier model

```python
bc = BaggingClassifier(base_estimator=dt,n_estimators=15,max_features=x.shape[1],random_state=25)
bc.fit(x_train,y_train)
```

```python
mscore(bc)
```

```python
ypred_bc = bc.predict(x_test)
print(ypred_bc)
```

```python
eval_metrics(y_test,ypred_bc)
```

### Checking Best Model

```python
best_model = pd.DataFrame({'Model Name':['LogisticRegression','KNeighborsClassifier','DecisionTreeClassifier','AdaBoostClassifier','BaggingClassifier'],
                          'Training Score':[lr.score(x_train,y_train),knn.score(x_train,y_train),dt.score(x_train,y_train),ada.score(x_train,y_train),bc.score(x_train,y_train)],
                          'Testing Score':[lr.score(x_test,y_test),knn.score(x_test,y_test),dt.score(x_test,y_test),ada.score(x_test,y_test),bc.score(x_test,y_test)]})
best_model
```
